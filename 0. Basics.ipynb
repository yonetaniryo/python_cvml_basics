# Function and Class in Python

Defining and executing functions

```python
import numpy as np

# function
def my_func(a, b):
    return a + b

a = 1
b = 2
print(my_func(a, b))

a = np.random.rand(2)
b = np.random.rand(2)
print(my_func(a, b))
```

Defining and executing classes

```python
class my_class():
    def __init__(self, a, b):
        self.a = a
        self.b = b
    
    def add(self):
        return self.a + self.b
    
    def sub(self, c):
        return self.a - c
    
cls = my_class(1, 2)
print(cls.add())
print(cls.sub(5))
```

# Plotting with `matplotlib`

```python
# plotting
import matplotlib.pyplot as plt
%matplotlib inline

t = np.linspace(0, np.pi, 100)
sint = np.sin(t)
cost = np.cos(t)

plt.figure()
plt.plot(t, sint)
plt.plot(t, cost)
```

You can tile multiple images with `plt.subplot`.

```python
# subplotting
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(t, sint)
plt.subplot(1, 2, 2)
plt.plot(t, cost)
```

You can draw scatter plots with `plt.scatter`.

```python
# scatter plotting
plt.figure()
plt.scatter(sint, cost)
```

# Loading and Showing Images and Videos

Let's load images and videos using `scikit-image (skimage)` and `scikit-video (skvideo)`. You can show loaded images with `plt.imshow`.

```python
# loading image
from skimage.io import imread
img = imread('data/cat.jpg')
plt.imshow(img)
```

```python
# loading image
# !pip install scikit-video
from skvideo.io import vread
video = vread('data/yks_cvpr2016_sub/pointing/00000_f.mp4')

for t in range(30):
    plt.subplot(5, 6, t + 1)
    plt.imshow(video[t * 2])
    plt.axis('off')
    plt.title(t, fontsize=8)
```
