# Pythonにおける関数とクラス

関数の定義と実行

````python
import numpy as np

# function
def my_func(a, b):
    return a + b

a = 1
b = 2
print(my_func(a, b))

a = np.random.rand(2)
b = np.random.rand(2)
print(my_func(a, b))
````

クラスの定義と実行

````python
class my_class():
    def __init__(self, a, b):
        self.a = a
        self.b = b
    
    def add(self):
        return self.a + self.b
    
    def sub(self, c):
        return self.a - c
    
cls = my_class(1, 2)
print(cls.add())
print(cls.sub(5))
````

# Matplotlibによるプロット

````python
# plotting
import matplotlib.pyplot as plt
%matplotlib inline

t = np.linspace(0, np.pi, 100)
sint = np.sin(t)
cost = np.cos(t)

plt.figure()
plt.plot(t, sint)
plt.plot(t, cost)
````

```plt.subplot```で複数の画像を並べることもできます．

````python
# subplotting
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(t, sint)
plt.subplot(1, 2, 2)
plt.plot(t, cost)
````

散布図は```plt.scatter```で描画できます．

````python
# scatter plotting
plt.figure()
plt.scatter(sint, cost)
````

# 画像，映像の読み込みと表示

ここでは，scikit-image (```skimage```)とscikit-video (```skvideo```) を用いて画像，映像を読み込みましょう．読み込んだ結果は```plt.imshow```で表示できます

````python
# loading image
from skimage.io import imread
img = imread('data/cat.jpg')
plt.imshow(img)
````

Scikit-videoによる映像の読み込みと表示

````python
# loading image
# !pip install scikit-video
from skvideo.io import vread
video = vread('data/person15_walking_d1_uncomp.avi')

for t in range(30):
    plt.subplot(5, 6, t + 1)
    plt.imshow(video[t * 2])
    plt.axis('off')
    plt.title(t, fontsize=8)
````
